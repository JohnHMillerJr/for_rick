# Comparing MODS Reponse Curves

---


## Steps:

- Reads "MODS Blue Response Curves" \_resp1d.fits files from "./sensFunc/resp1d/\*.fits."
- Plots all (raw or smooth) response curves with "mean" response curve.
- Picks a fiducial response curve, finds the response difference at a normalizing wavelength for each night, then shifts the response to the fiducial, then calculates the residuals from that fiducial.

```python
# GENERAL IMPORTS
import os
import glob
import numpy as np
import pandas as pd

# PLOTTING IMPORTS
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from scipy import interpolate

# FITS IMPORT
from astropy.io import fits

# DATE IMPORTS
import matplotlib.dates as mdates
import datetime as dt 
```

## General Declarations for Plotting

```python
# DECLARATIONS

# graph bounds
minWave = 3300.0
maxWave = 5700.0

# graphing prefs
aspect = 4.0/3.0
textWidth = 6.0 # inches
figFmt = 'png'
dpi = 600
plotWidth = dpi*textWidth
plotHeight = plotWidth/aspect
wInches = textWidth 
hInches = wInches/aspect 
lwidth = 0.5

# global plotting params
plt.rcParams.update({'font.size': 10, 'xtick.labelsize' : 8,
                    'ytick.labelsize' : 8, 'lines.linewidth': 0.5})

# sets up graph
def setGraph(ax):

    # tick params
    ax.tick_params('both',length=6,width=lwidth,which='major',direction='in',top='on',right='on')
    ax.tick_params('both',length=3,width=lwidth,which='minor',direction='in',top='on',right='on')

    # x ticks & label
    ax.xaxis.set_major_locator(MultipleLocator(500))
    ax.xaxis.set_minor_locator(MultipleLocator(100))

    # y ticks & label
    ax.yaxis.set_major_locator(MultipleLocator(0.5))
    ax.yaxis.set_minor_locator(MultipleLocator(0.1))
```

## Aquires Response Curves & Data

The response curves are taken from a local directory: './sensFunc/resp1d/'. This directory is the output of \_resp1d.fits files from the "MODS Blue Response Curves" Jupyter notebook. The program scrapes for all .fits files in the directory.



## Data Summary

---

`data`

`data[:][0] = date` \
`data[:][1] = wavelengths` \
`data[:][2] = response` \
`data[:][3] = pixel shift`

---

There are raw data arrays for each mod#b and std G191-B2B and GD153.

```python
# skips
skip = True
if skip: 
    g191_skips = ['20200920', '20220228']; gd153_skips = []
else:
    g191_skips = []; gd153_skips = []

    
# read in function
def readData(allFiles, skips):
        
    # print(len(allFiles))
    
    # either raw or smooth response - - -
    raw = False
    global label
    if raw: 
        respType = 'raw_resp'
        label = ' Raw R'
    else: 
        respType = 'response'
        label = ' Smooth R'

    # reading in data
    refWave = []; num = 0
    data = np.empty( shape=(len(allFiles) - len(skips), 6), dtype='object')
    for i, file in enumerate(allFiles):

        i -= num # for skips

        # data
        resp1d = fits.open(file, memmap=False)
        wave = resp1d[1].data['wavelength']
        resp = resp1d[1].data[respType]
        pixShift = resp1d[0].header['PIXSHIFT']
        date = file.split('_')[2]

        if date in skips: 
            num += 1
            continue

        if i == 0: refWave = wave

        splf = interpolate.InterpolatedUnivariateSpline(wave,resp)
        newResp = splf(refWave) # resampling

        # saving
        data[i][0] = date
        data[i][1] = wave
        data[i][2] = newResp
        data[i][3] = pixShift

        resp1d.close()
        
    return data


# getting files - - -
p1d_dir = './sensFunc/resp1d/'
g191_1b_files = glob.glob(p1d_dir+'*%s*m%i%s*.fits' % ('G191', 1, 'b'))
g191_2b_files = glob.glob(p1d_dir+'*%s*m%i%s*.fits' % ('G191', 2, 'b'))
gd153_1b_files = glob.glob(p1d_dir+'*%s*m%i%s*.fits' % ('GD153', 1, 'b'))
gd153_2b_files = glob.glob(p1d_dir+'*%s*m%i%s*.fits' % ('GD153', 2, 'b'))

# reading in data - - - - - -
g191_mod1b_raw = readData(g191_1b_files, g191_skips)
g191_mod2b_raw = readData(g191_2b_files, g191_skips)
gd153_mod1b_raw = readData(gd153_1b_files, gd153_skips)
gd153_mod2b_raw = readData(gd153_2b_files, gd153_skips)
rawData = [g191_mod1b_raw, g191_mod2b_raw, gd153_mod1b_raw, gd153_mod2b_raw]

# print(gd153_2b_files_raw)
```

```python
#> helpful functions

# returns the index with the lambda closest to the given lambda
def waveIndex(waves, normLam):
    
    adjust = [ abs(x-normLam) for x in data[0][1] ]
    index = np.argmin(adjust)
    
    return index

# returns avg resp at given lambda with width
def avgRespAtWave(waves, resp, normLam, normWid):
    
    avgResp = []
    for j, wave in enumerate(waves): # appends resp if wave in range
        if abs(wave-normLam) <= (normWid/2): avgResp.append(resp[j])
    
    return np.mean(avgResp)
```

Gettings the avg. resp. at specific wavelengths for each mod#b and std star. \
And setting up dates for plotting.

```python
# setting up resp @ lambda dataframes
g1Mod1 = pd.DataFrame(); g1Mod1['date'] = g191_mod1b_raw[:,0]
g1Mod2 = pd.DataFrame(); g1Mod2['date'] = g191_mod2b_raw[:,0]
gd1Mod1 = pd.DataFrame(); gd1Mod1['date'] = gd153_mod1b_raw[:,0]
gd1Mod2 = pd.DataFrame(); gd1Mod2['date'] = gd153_mod2b_raw[:,0]
data = [g1Mod1, g1Mod2, gd1Mod1, gd1Mod2]

# setting up dates + getting avg. resp @ lambdas
for i, std_mod in enumerate(rawData):
    
    ds = [] # date set up
    for date in std_mod:
        year = int(date[0][:4])
        month = int(date[0][4:6])
        day = int(date[0][6:8])
        d = dt.datetime(year, month, day)
        ds.append(d)
    data[i]['d'] = ds
    
    waveWid = 10.0 # avg. resp @ lambda
    waves = np.arange(3500, 6000, 500)
    for wave in waves:
        avgResp = []
        for date in std_mod:
            avgResp.append(avgRespAtWave(date[1], date[2], wave, waveWid))
        data[i][str(wave)] = avgResp

g1Mod1.head()
```

# Overplotting Response Curves

---

`mean`

`mean[0] = wavelengths` \
`mean[1] = mean response curve` \
`mean[2] = std` 

---

```python
#> NEED TO SELECT DATA SET
data_set = 0
rawData = [g191_mod1b_raw, g191_mod2b_raw, gd153_mod1b_raw, gd153_mod1b_raw]
selData = rawData[data_set]

# getting mean response curve - - -
waves = selData[0][1] # reference wavelengths -- assumes all same
mean = np.zeros( shape=(4, len(waves)), dtype='object')
for i, wav in enumerate(waves):
    
    # gets all f_lambda's
    vals = []
    for j in range(len(selData)): 
        try:
            mean[1][i] += selData[j][2][i]
        except:
            print(selData[j])
        vals.append(selData[j][2][i])
    
    # getting mean + saving wav
    mean[2][i] = np.std(vals)
    mean[1][i] /= len(selData)
    mean[0][i] = wav

    
# getting graphing bounds
minR = np.min([np.min(x) for x in selData[:,2]])
maxR = np.max([np.max(x) for x in selData[:,2]])
dR = maxR - minR
minResp = minR - 0.05*dR
maxResp = maxR + 0.05*dR


# initializing plot
fig, ax = plt.subplots(2,1,figsize=(wInches,hInches),dpi=dpi,
                             gridspec_kw={'height_ratios':[2.5,1],'hspace':0})
titles = [r'G191-B2B, MODS1B: Response Curves', r'G191-B2B, MODS2B: Response Curves',
         r'GD153, MODS1B: Response Curves', r'GD153, MODS2B: Response Curves']
out = ['g191-1b', 'g191-2b', 'gd153-1b', 'gd153-2b']

for i, axes in enumerate(ax.flat):
    
    # main graph
    if i == 0: 

        # title
        axes.set_title(titles[data_set])
        axes.set_xlabel(r'Wavelength [$\AA$]')
        axes.set_ylabel(r'Response$_\lambda$ [erg/s/cm$^2$/$\AA$]')

        setGraph(axes) # sets up graph

        # plotting!
        labels = np.full(len(selData), '_nolegend_')
        labels[0] = str(len(selData))+label
        refWave = []
        for j, date in enumerate(selData):
            if j == 0: refWave = date[1]
            axes.step(refWave,date[2],'-',zorder=9,label=labels[j], alpha=1, color='0.8')
        axes.step(mean[0],mean[1],'-',lw=0.5,zorder=9,label='Mean', alpha=1, color='r')
        axes.legend(fontsize=5)
    
    # standard deviation
    if i == 1:
        axes.set_ylabel('Std')
        axes.grid(ls=':', alpha=0.3)
        
        for j in range(len(selData)): 
            axes.step(mean[0], mean[2], color='r', alpha=1)

plt.savefig('./plots/all-resp-%s.pdf' % out[data_set], dpi=5, bbox_inches='tight')
plt.show()
```

## Declaring References, Norms, etc.

```python
# comparisons lambda & width
normLam = 4500.0
normWid = 10.0 # total
print('> Normalization Wavelength: %.1f A' % normLam)
print('> Normalization Width: %.1f A \n' % normWid)

# finding closest lambda index & response
# assumes wavelength grid is same for all
indexes = []
for data in rawData:
    adjust = [ abs(x-normLam) for x in data[0][1] ]
    indexes.append(np.argmin(adjust))


# finding reference (max at normLam)
refs = []
for j, raw in enumerate(rawData):
    
    maxResp = -np.inf
    for i, date in enumerate(raw):
        if date[2][indexes[j]] >= maxResp:
            maxResp = date[2][indexes[j]]
            ref = i 
    refs.append(ref)
    
# other refs for G191-B2B.mod1b
# ref = 26  # middle of 2022
ref = 47  # 

# printing refs
print('> G191.MOD1B Reference: %s, %i' % (rawData[0][refs[0]][0], refs[0]))
print('> G191.MOD2B Reference: %s, %i' % (rawData[1][refs[1]][0], refs[1]))
print('> GD153.MOD1B Reference: %s, %i' % (rawData[2][refs[2]][0], refs[2]))
print('> GD153.MOD2B Reference: %s, %i' % (rawData[3][refs[3]][0], refs[3]))
```

## Stacking Resp. Curves

```python
#> NEED TO SELECT DATA SET
data_set = 0
rawData = [g191_mod1b_raw, g191_mod2b_raw, gd153_mod1b_raw, gd153_mod1b_raw]
data = [g1Mod1, g1Mod2, gd1Mod1, gd1Mod2]
selData = rawData[data_set]; pdData = data[data_set]


# initializing plot
fig, ax = plt.subplots(2,1,figsize=(wInches,hInches),dpi=dpi,
                             gridspec_kw={'height_ratios':[2.5,1],'hspace':0})
titles = [r'G191-B2B, MODS1B: Response Curves', r'G191-B2B, MODS2B: Response Curves',
         r'GD153, MODS1B: Response Curves', r'GD153, MODS2B: Response Curves']
out = ['g191.1b', 'g191.2b', 'gd153.1b', 'gd153.2b']

# plotting!
for i, axes in enumerate(ax.flat):
    
    axes.grid(ls=':', alpha=0.3)

    # main graph
    if i == 0: 
        
        # title
        axes.set_title(titles[data_set])
        axes.set_xlabel(r'Wavelength [$\AA$]')
        axes.set_ylabel(r'Response$_\lambda$ [erg/s/cm$^2$/$\AA$]')

        setGraph(axes) # sets up graph
        
        refWave = []
        for j, date in enumerate(selData):
            
            # shifting
            date_resp = pdData[ pdData['date'] == date[0]][str(int(normLam))].to_numpy()[0]
            ref_resp = pdData[ pdData['date'] == selData[refs[data_set]][0]][str(int(normLam))].to_numpy()[0]
            shift = date_resp - ref_resp
            resp = [x - shift for x in date[2]]

            # plotting
            if j == 0: refWave = date[1]
            if j == refs[data_set]: axes.step(refWave, resp,'-', zorder=9, label=date[0], alpha=1, c='r')
            else: axes.step(refWave, resp,'-', zorder=0, alpha=0.5, c='0.8')
            
        # legend
        axes.legend(fontsize=5)
        
        # plotting normLam & normWid
        ylims = axes.get_ylim()
        axes.axvline(normLam, c='k', alpha=0.4)
        axes.fill_betweenx([0, -10], [normLam-(normWid/2)], [normLam+(normWid/2)], color='0.8', alpha=0.5, linewidth=0)
        axes.set_ylim(ylims)
    
    # residuals
    if i == 1:
        
        axes.set_ylabel('Fractional Resid.')
        axes.grid(ls=':', alpha=0.3)
        axes.set_ylim(-0.15,0.15)
        
        for j, date in enumerate(selData):

            # shifting
            date_resp = pdData[ pdData['date'] == date[0]][str(int(normLam))].to_numpy()[0]
            ref_resp = pdData[ pdData['date'] == selData[refs[data_set]][0]][str(int(normLam))].to_numpy()[0]
            shift = date_resp - ref_resp
            resp = [x - shift for x in date[2]]
            
            axes.step(refWave, (resp - selData[refs[data_set]][2]) / selData[refs[data_set]][2] ,'-',zorder=9, alpha=0.5, c='0.8')
            
        # plotting normLam & normWid
        ylims = axes.get_ylim()
        axes.axvline(normLam, c='k', alpha=0.4)
        axes.fill_betweenx([-1, 1], [normLam-normWid], [normLam+normWid], color='0.8', alpha=0.5, linewidth=0)
        axes.set_ylim(ylims)

plt.savefig('./plots/stacked-resp-%s.pdf' % out[data_set], dpi=5, bbox_inches='tight')
plt.show()
```
